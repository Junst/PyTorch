```python
import torch
a = torch.randn(3, 4)
a.transpose_(0, 1)
b = torch.randn(4, 3)
```

```python
# a 텐서 메모리 주소 예시
for i in range(4) :
	for j in range(3) :
		print(a[i][j].data_ptr())

# b 텐서 메모리 주소 예시
for i in range(4):
    for j in range(3):
        print(b[i][j].data_ptr())
```
